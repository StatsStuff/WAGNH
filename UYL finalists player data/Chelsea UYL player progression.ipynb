```python
import requests
import pandas as pd

from tqdm import tqdm
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv("uyl_player_data.csv", index_col=0)
```

```python
chels = df[df.uyl_team == 'Chelsea FC']
```

```python
pd\
   .pivot_table(chels, values='minutes', index='uyl_name', columns='competition', aggfunc='sum', fill_value=0, margins=True) \
   .sort_values('All', ascending=False) \
   .drop('All', axis=1) \
   .drop('All', axis=0) \
   .plot.bar(xlabel="player", ylabel="total minutes", figsize=(8,8), stacked = True, title='How many minutes in total have Chelsea UYL finalists played in international cups?')
pass
```

```python
pd\
   .pivot_table(chels, values='minutes', index='uyl_name', columns=['club'], aggfunc='sum', fill_value=0, margins=True) \
   .sort_values('All', ascending=False) \
   .drop('All', axis=1) \
   .drop('All', axis=0) \
   .plot.bar(xlabel="player", ylabel="total minutes", figsize=(8,8), stacked = True, title='For which teams have Chelsea UYL finalists played in international cups?')
pass
```

```python
pd\
   .pivot_table(chels[chels.club == 'Chelsea FC'], values='minutes', index='uyl_name', columns='competition', aggfunc='sum', fill_value=0, margins=True) \
   .sort_values('All', ascending=False) \
   .drop('All', axis=1) \
   .drop('All', axis=0) \
   .plot.bar(xlabel="player", ylabel="total minutes", figsize=(8,8), stacked = True, title='How many minutes in total have Chelsea UYL finalists played in international cups for Chelsea?')
pass
```
